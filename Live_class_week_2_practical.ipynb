```python
#importing pandas library for data manipulation
import pandas as pd
cars=pd.read_csv("mtcars.csv")  #assigning variable to read the dataset 
cars #printing the file
```

```python
cars.head(10) # printing first 10 rows of dataset
```

```python
cars.tail(10) #displays last 10 rows
```

```python
cars.info() #shows different data types of the data set and the null values present(if any)
```

```python
cars=cars.rename(columns={'mpg':'mileage'}) #renaming columns
cars
```

```python
cars.qsec=cars.qsec.fillna(cars.qsec.mean()) #filling null valus with mean if any null values are there
cars
```

```python
cars=cars.drop(columns=['disp']) #dropping column disp
cars
```

```python
import matplotlib.pyplot as plt #importing matplotlib
%matplotlib inline
y= cars['hp'] #y-axis will have hp column of dataset
x=range(32) #range from 0-32 in x axis , as we have a total of 32 rows
plt.plot(x,y) # printing the graph, plot by default prints line chart if we dont specify the type of plot





```
