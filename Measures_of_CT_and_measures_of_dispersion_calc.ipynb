```python
import pandas as pd
import numpy as np #Library for performing statistical calculations.
Med= pd.read_csv("MedGPA.csv")
Med
```

```python
GPA= Med["GPA"]
GPA #Defining GPA from file to perform operations on it.
```

###Mean


```python
GPA.mean()
```

###Mode

```python
GPA.mode()
```

###Median

```python
GPA.median()
```

Interquartile Range

```python
#calculating IQR
q3,q1= np.percentile(GPA,[75,25]) #GPA is the column name
iqr=q3-q1
iqr #printing iqr
```

Variance

```python
GPA.var()
```

Standard Deviation

```python
GPA.std()
```

Summary

```python
GPA.describe() #to describe the summary of the data calculations
```

###Bar plot

```python
height=Med['Sex'].value_counts() #function to count total nos in each category
height #checking the nos.
sex=['Feamle', 'Male'] #creating categories based on nos.
plt.bar(sex,height) #creating plot, sex is for x-axis , height is for y-axis
plt.show() #showing the plot
```

Pie chart

```python
accept=Med['Accept'].value_counts() #counting total no of values in each category
accept #checking th nos.
```

```python
Category=['A','d'] #creating categories based on nos
fig=plt.figure(figsize=(7,5)) #10,7 is for the size of the figure it can be anything
plt.pie(accept,labels=Category) #accept-count, label=categories
plt.show() #for displaying chart
```
